```python
#Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
#criando dataframe
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
#visualizando dados
df.head(5)
```

```python
#Verificando total de receita 
round(df["Valor Venda"].sum(),2)#
```

```python
#Calculando o custo
df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"])
df.head()
```

```python
#custo total
round(df["Custo"].sum(),2)
```

```python
df.head(1)
```

```python
#Criando a coluna lucro
df["lucro"] = df ["Valor Venda"] -df["Custo"]
df.head(1)
```

```python
#soma do lucro
round(df["lucro"].sum(),2)
```

```python
#calculando o lucro medio
round(df['lucro'].mean(),2)
```

```python
df.head(1)
```

```python
#Criando a Coluna de ano venda
df['ano venda'] = df["Data Venda"].dt.year
df.head(5)
```

```python
round(df.groupby('ano venda')['lucro'].sum(),2)
```

```python
df.groupby(df['ano venda'])['lucro'].sum().plot.bar(title="VENDAS POR ANO")
plt.xlabel("ano venda")
plt.ylabel("lucro")
```
